## Introdução a estatística e amostragem <br>
##### A Estatística é a ciência que estuda os métodos de coleta, análise, organização, descrição, interpretação e apresentação de dados experimentais. Os dados são obtidos com o intuito de responder um problema do mundo real, utilizando amostras. A estatística subdivide-se em três grande áreas:
* Descritiva: Parte da estatística voltada para a organização e descrição dos dados (Organização, resumo e apresentação dos dados);
* Probabilidade: Útil para avaliar situações em que envolvem o acaso (Eventos aleatórios);
* Inferência: Refere-se a análise e interpretação dos dados (Apartir de uma amostra, obter conclusões sobre a população);

<br>

### Amostragem <br>
##### Amostra é todo subconjunto de unidades retiradas de uma população. O processo de retirada de informações dos "n" elementos amostrais, deve seguir um método criterioso e adequado que são os tipos de amostragem. Basicamente existem amostras probabilísticas e não probabilísticas:
* Probabilística: Todos os elementos da população apresentam probabilidade maior que zero de serem selecionados (Aleatória simples, estratificada, sistemática e por conglomerados)
* Não probabilística: Quando não há probabilidade clara/conhecida de seleção dos elementos. Os elementos são escolhidos de forma julgamental (Acidental, intencional, por cotas)

<br>

### Técnicas de amostragem probabilísticas <br>
* Na amostra aleatória simples (AAS) podem ser realizadas com ou sem reposição. No caso em que há reposição, cada elemento pode ser sorteado mais de uma vez. 
* Na amostra estratificada a população é dividida em estratos homogêneos (grupos com elementos de características comuns) e é selecionada um amostra aleatória de cada estrato.
<img  width="30%" src="_imagens/Amostragem estratificada.png">
* Na amostra sistemática os elementos são selecionados seguindo uma regra pré definida.
<img  width="30%" src="_imagens/Amostragem sistemática.png">
* Na Amostra por conglomerados a população (extensa) é divida em miniaturas da população (não homogêneas) e seleciona-se uma amostra aleatória desses conglomerados. Exemplo: Estado de São Paulo > Possuem "n" bairros distintos > Sorteia-se um bairro para se analisado

### Tipos de dados, tabelas e distribuição de frequência <br>
##### A estatística é uma parte de matemática aplicada que fornece métodos para coleta, organização, descrição, análise e interpretação de dados e para a utilização na tomada de decisões. 
##### Independente do método de obtenção (Amostragem ou censo), os dados precisam ser explorados e interpretados para atingir os resultados desejados.  <br>
##### A coleta, a organização, a descrição dos dados, o cálculo e a interpretação de coeficientes pertencem a estatística descritiva, enquanto a análise e a interpretação dos dados, associado a uma margem de incerteza, ficam a cargo da estatística indutiva ou inferencial, também chamada como a medida da incerteza ou métodos que se fundamentam na teoria da probabilidade. <br>
##### A exploração desses dados consiste em organiza-los em tabelas e gráficos, disponíveis para cada tipo específico de dados: qualitativos e quantitativos.

<br>

### Tipo de dados <br>
##### Ao coletar dados precisamos entender qual é o tipo da variável analisada
* Categorizados/Categoricos (Qualitativos)
    * Nominais (Nomes não seguem uma ordem)
    * Ordinais (Nomes que seguem uma ordem)
* Numéricos (Quantitativos)
    * Discretos (Um valor inteiro dentro de um intervalo)
    * Contínuos (Um valor flutuante dentro de um intervalo)
    
 <br>
 
### Construção de tabela <br>
##### Para facilitar a análise dos dados condensamos os dados em forma de tabela. Uma tabela é um quadro que resume um conjunto de observações. A tabela tem que ter:
* Título: Explica o conteúdo
* Corpo: Conjunto de linhas e colunas
* Cabeçalho: Especifica o conteúdo das colunas
* Coluna indicadora: Especifica o conteúdo das linhas
* Casa ou células: Espaço destinado a um só número, interseção de uma linha e uma coluna
* Linha: Série de células organizadas horizontalmente
* Coluna: Série de células organizadas verticalmente
* Rodapé: Pode conter fontes, notas ou chamadas
 
<br>
 
### Tabelas de distribuições de frequência <br>
##### Podemos melhorar a interpretação dos dados transformando a tabela numa distribuição de frequêcia
* Frequência absoluta: Quantidade de ocorrências de um valor
* ntervalo de classes: Agrupando os valores em intervalos que possui mesma amplitude
* Frequência relativa: Frequência absoluta / Total x 100
* Ponto médio: Valor médio do intervalo de classes (Máximo - Mínimo / 2)
* Frequência acumulada: Acumulando as frequências absolutas
* Frequência relativa acumulada: Acumulando as frequências relativa

### Visualizações gráficas <br>
##### Os gráficos são representações que facilitam a análise de dados. As informações em um gráfico devem estar dispostas de maneira simples, clara e verídica. Para isso alguns elementos são fundamentais como título, eixos, legenda, rótulos e linhas.
* Gráfico de linhas: Utilizado para demonstrar uma sequência numérica de um certo dado ao longo do tempo. É indicado para demonstrar evoluções ou regressões que ocorrem em sequência para que o comportamento dos fenomenos e suas transformações sejam observadas.
* Gráfico de coluna: Composto por duas linhas ou eixos, um vertical e outro horizontal. 
* Gráfico de barras: As barras podem ficar na horizontal ou na vertical. 
* Gráfico de setores/pizza: É empregado sempre que desejamos ressaltar a participação do dado (apenas uma variável) no total. Utilizado com até 5 categorias da variável.  
* Histograma: É proporcional a soma das frequências simples ou absolutas.
* Diagrama de caixas / boxplot: É uma maneira gráfica de representar a alteração dos dados de uma variável por meio de quartis.
    * Limite inferior
    * Limite superior
    * 1° quartil = 25%
    * 2° quartil = Mediana = 50%
    * 3° quartil = 75%
    * 4° quartil = 100%
* Gráfico de dispersões: São utilizados na análise do relacionamento entre duas variáveis

### Medidas de posição e medidas de dispersão <br>
##### Para entender as caracterísitcas gerais de uma amostra precisamos fornecer um resumo que seja representativo, usando apenas um valor.
* Medidas de posição ou endência central (Dão ideia do centro em torno do qual os dados se distribuem): 
    * Moda (Valor que aparece com maior frequência) = Amodal, Modal ou Bimodal
    * Média (Somatório dos valores e divido pela sua quantidade) = A média é influenciada por valores discrepantes
    * Mediana (Valor central dos valores ordenados)
    * Quartis (Dividi os valores em 4 partes iguais = 25% para cada)
    * Percentis (Dividi os valores em 100 partes iguais)
* Medidas de variabilidade ou dispersão (Revelam o quanto se dispersam):
    * Mínimo
    * Máximo
    * Amplitude total (Maior valor - Menor valor) = A amplitude é influenciada por valores discrepantes
    * Variância = Desvio padrão²
    * Desvio padrão (Quanto maior variação do desvio padrão em relação a média)

### Coeficiente de correlação de Pearson (r)<br>
##### É um grau de relação/associação entre duas variáveis quantitativas e indica o grau da intensidade da correlação, além do sentido dessa correlação (positivo ou negativo). 
* Os limites de r são -1 e +1.
* se r = +1, há uma correlação perfeita e positiva entre as variáveis.
* se r = -1, há uma correlação perfeita e negativa entre as variáveis.
* se r = 0, ou não há correlação entre as variáveis ou a relação que porventura exista, não é linear.

### Regressão Linear Simples <br>
##### Quando analisamos a relação entre duas variáveis, determinamos a correlação entre elas, ou seja, como e o quanto essas variáveis quantitativas estão relacionadas. <br>
##### Análise de regressão estuda determinada variável em função de outra. O seu objetivo é descrever, através de um modelo matemático, a relação entre duas variáveis a partir de N observações (e também fornecer um método de previsão que permita a avaliação em termos de confiabilidade, permitindo a transformação da incerteza em risco) <br>
##### A variável sobre a qual desejamos fazer uma estimativa recebe o nome de variável dependente e a outra recebe o nome de variável independente. <br>
##### A dependência de uma variável, que indicaremos por Y, em relação a outra variável, que indicaremos por X, é estudada, em Estatística, por uma técnica denominada regressão. <br>

<img  width="30%" src="_imagens/Regressão linear.png">

<br>

##### Equação da reta da regressão linear: Y = aX + b (Explica a relação funcional entre a variável independente Y e a variável dependente X)
* a = inclinação ou coeficiente angular 
* b = coeficiente linear

<img  width="30%" src="_imagens/Coeficiente angular e coeficiente linear.png">

* Média de Y = Somatória dos valores de Y dividido por N
* Média de X = Somatória dos valores de X dividido por N

<br>

### Coeficiente de determinação (r²) <br>
##### É matematicamente igual ao coeficiente de correlação de Pearson ao quadrado, por tanto, variam entre 0 e 1. Mostra a força da relação linear entre X e Y.  <br>
##### Se a reta de regressão passa exatamente sobre os pontos do diagrama de dispensão, r² = 1. Nesse caso a correlação linear é perfeita. Toda variação em Y está relacionada com a variação em X. <br>
##### Se os valores de X aumentam e Y varia ao acaso, não existe relação linear entre as variáveis, assim, o valor do coeficiente de determinação r² será zero ou bem proximo de zero.<br>
##### O coeficiente de determinação pode ser entendido como a proporção de variação de uma variavel que é explicada pela variação da outra variavel. Mede-se assim, a contribuição de X na previsão de Y. <br>
##### Significa que r² (%) do total da variação de Y pode ser explicado enquanto (100%-r²) o resto da variação constitui um resíduo, ou seja, não são explicados pela equação da reta.

### Introdução a probabilidade <br>
##### Um modelo matemático ajudará a investigar de maneira bastante precisa esse fenômeno/experimentos aleatórios. Encontramos na natureza dois tipos de fenomenos:
* Determinísticos: Os resultados são sempre os mesmos e determinados pelas condições sob as quais o procedimento seja executado. 
* Não determinísticos / Probabilísticos / Aleatório : São aplicados em situações que envolvem incerteza. Os resultados variam de uma observação para outra. As condições do experimento determinam apenas o comportamento probabilístico do resultado observável. 

<br>

##### Fenômenos aleatórios, onde situações ou acontecimentos cujos resultados não podem ser previstos com certeza, dependem do acaso. Quando o experimento é executado repetidamente, sob as mesmas condições, os resultados individuais parecem ocorrer de uma forma casual (acidental). No entanto, a medida que o número de repetições aumenta, surgem certos padrões na frequência de ocorrência dos resultados. É esta regularidade (ou padrão) que torna possível construir um modelo matemático para analisar o experimento. <br>
* Espaço amostral (S): É uma lista com todos os resultados possíveis/possibilidade de um experimento. Uma forma de determinar o espaço amostral é atraves do Diagrama de Venn, em que o espaço amostral é representado por retângulos e os eventos por círculos. Exemplo: 
    * Experimento = Lançamento de um dado
    * S = {1,2,3,4,5,6}
* Evento (E): É qualquer subconjunto do espaço amostral. Cada evento tem determinada característica, que o distingue dos outros. Classificação: Evento impossível, Evento certo (Resultado igual ao espaço amostral), Evento elementar (Um resultado) e Evento composto (Mais de um resultado). Exemplos: 
    * Evento: Resultado ser par = {2,4,6} 
    * Evento: Resultado é maior 3 = {4,5,6}
* Utilizamos probabilidade para quantificar quão provável é determinado evento.

<br>

### Combinação de eventos 
##### Se A e B são eventos de um mesmo espaço amostral, diremos que ocorre o evento.
* A união B, A soma B ou A mais B, se e somente se A ocorre ou B ocorre;
* A produto B, A vezes B ou A interseção B, se e somente se A ocorre e B ocorre simultaneamente. 
* Complementar de A (não A) se e somente se A não ocorre.
* A menos B, A diferença B, se e somente se A ocorre e B não ocorre.
* Dois eventos A e B são mutuamente excludentes (exclusivos) se não puderem ocorrer juntos. Resultado = 0

### Probabilidade <br>
##### Na natureza, existem fenômenos probabilísticos, também conhecidos como aleatórios, que envolvem incerteza, os resultados variam de uma observação para outra. Porém, esses fenômenos, ao serem repetidos inúmeras vezes, apresentam regularidade em termos de frequência. A probabilidade é uma medida com a qual podemos esperar a chance de ocorrência de um determinado evento, atribuido um número (valor) entre 0 e 1. <br>

* Definição clássica diz que a probabilidade de ocorrer um evento com determinada característica de interesse é dada pelo quociente entre o número de eventos favoráveis e o número de eventos possíveis. >>> P(A) = n° de resultados do evento A / n° total de resultados no espaço amostral = n(A) / n(S)
* Definição frequentista considera o limite de frequências relativas como o valor de probabilidade. Se um experimento aleatório é repetido um número grande de vezes, N, e seja nA o número de ocorrências do evento A, a probabilidade de A é dada por >>> fn(A) = frequência do evento A / total de realizações = n(A) / n
* Quando n tende a infinito, fn(A) se aproxima de P(A)

<br>

#### Tipos de agrupamentos <br>
##### A probabilidade de um evento A é definida como sendo um número P(A) satisfazendo os axiomas: 

<img  width="40%" src="_imagens/Axiomatica.png">

##### Eventos independentes: Dizemos que dois eventos são independentes quando a realização ou não realização de um dos eventos não afeta a probabilidade da realização do outro e vice-versa. Dois eventos são independentes se e somente se P(A interseção B) = P(A) * P(B) <br>
##### Eventos mutuamente exclusivos (excludentes) ou eventos disjuntos: Dizemos que dois eventos são disjuntos ou mutuamente excludentes quando a realização de um exclui a realização do(s) outro(s). Assim, no lançamento de uma moeda, o evento "tirar cara" e o evento "tirar coroa" são mutuamente excludentes, já que, ao se realizar um deles, o outro não se realiza. > P(A interseção B) = 0 <br>
##### Probabilidade condicional é um dos conceitos mais imporrtantes da teoria de probabilidades, pois frequentemente se quer calcular seu valor quando se tem alguma informação parcial a respeito do resultado de um experimento aleatório. Além disso, mesmo quando não se tem essa informação parcial, as probabilidades condicionais muitas vezes são utilizadas para computar mais facilmente valores de probabilidades que se tem interesse.  Sejam A e B dois eventos de um espaço amostral S, com P(A)>0. A probabilidade condicional de B, dado A, denotada por: P(B/A) = P(A interseção B) / P(A) 

### Análise Combinatória <br>
##### Estuda os problemas de contagem. Ela surgiu da necessidade de se calcular o número de possibilidades que podem ocorrer numa certa experiência, sem precisar descrever cada uma dessas possibilidades. É fundamental para dimensionar o espaço amostral. 
* Princípio fundamental da contagem (PFC): Na análise combinatória, o princípio fundamental da contagem consiste em separar o problema em etapas. O número total de possibilidades vai ser o produto das possibilidades em cada etapa. Exemplo: Uma senha formada por um número de 1 a 9 e uma letra de A a Z. Teremos: 
    * Número = 9 possibilidades
    * Letras = 26 possibilidades
    * 9 x 26 = 234 possibilidades

<br>

#### Tipos de agrupamentos <br>
##### Arranjo: Para calcular o número de possibilidades para forma um grupo ordenardo de P elemtnos, dentre N elementos disponiveis. 
* An,p = n! / (n-p)! 
##### Permutação: São agrupamentos ordenados, onde o número de elementos (n) do agrupamento é igual ao número de elementos disponíveis. Note que a permutação é um caso especial de arranjo quando o número de elementos é igual ao número de agrupamentos. Dessa forma, o denominador na fórmula do arranjo é igual a 1 na permutação. 
* Permutação simples: Pn = n!
* Permutação com repetição é uma permutação em que alguns dos elementos são iguais, ou se repetem algumas vezes. Para permutação de n elementos, em que um dos elementos se repete a vezes, outro b vezes e um terceiro c vezes: Pn = n! / a! * b! * c! 
##### Combinação: Para formar grupos não ordenados de p elementos, dentre os n disponíveis. A principal diferença para o arranjo é que na combinação a ordem não importa. 
* Cn,p = n! / p! * (n - p)!

<br>

#### Probabilidade de análise combinatória <br>
##### Permite analisar ou calcular as chances de obter determinado resultado diante de um experimento aleatório. A probabilidade é determinada pela razão entre o número de eventos possíveis e número deeventosfavoráveis, sendo apresentada pela seguinte expressão: P(A) = n(A) / n(S)
* P(A): Probabilidade de ocorrer um evento A
* n(A): n° de resultados do evento A (resultados favoráveis)
* n(S): n° total de resultados no espaço amostral (resultados possíveis)

### Probabilidade total, parcial e Teorema de Bayes <br>
##### Para relacionar probabilidades entre dois ou mais eventos ou achar novas probabilidades a partir das que temos, contamos com as fórmulas de probabilidade condicional, probabilidade total e teoremas de bayes. <br>
##### Teorema da probabilidade total diz que a probabilidade de um evento B ocorrer é a soma das probabilidades das intersecções do evento B com as partições A de outro evento A. 
* P(B) = Soma P(A interseção B) = Soma P (B/A) * P(A)
##### Teorema de Bayes deriva da probabilidade condicional. Ele relaciona as probabilidades condicionais entre dois eventos da seguinte forma:
* P(A/B) = P(B/A) * P(A) / P(B) ou 
* P(A/B) = P(B/A) * P(A) / Soma P (B/A) * P(A)
